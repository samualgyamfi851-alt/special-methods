```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)
```

```python
class Vector:
    def __init__(self, x):
        self.x = x

    def __sub__(self, other):
        return Vector(self.x - other.x)

    def __str__(self):
        return str(self.x)

v1 = Vector(10)
v2 = Vector(3)

print(v1 - v2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return Number(self.value * other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(5)
n2 = Number(4)

print(n1 * n2)
```

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __getitem__(self, index):
        return self.data[index]

g = Grades()

print(g[0])
print(g[1])
```

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)

print(double(10))
```

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.limit:
            raise StopIteration

        value = self.current
        self.current += 1
        return value

for num in Counter(5):
    print(num)
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

account = BankAccount(500)

if account:
    print("Account is active")
```
